# 作業：第二題
## 年份：2021年
[氣溫來源](https://www.cwb.gov.tw/V8/C/C/Statistics/monthlydata.html)

```python
import matplotlib.pyplot as plt
```

```python
TPE_max = [29.8, 28.4, 30.4, 30.9, 38.2, 37.0, 37.8, 35.1, 35.4, 33.8, 29.7, 27.2]
TPE_min = [6.8, 11.9, 11.9, 15.9, 19.8, 21.8, 24.9, 24.3, 23.7, 19.1, 13.9, 12.3]
month = [x for x in range(1,13)]
```

```python
# 設定長寬
plt.figure(figsize=(10,6), linewidth = 2)

# 設定圖樣 
plt.plot(month,TPE_max, 'r-o', label="Maximum Temperature")
plt.plot(month,TPE_min, 'g-s', label="Minimum Temperature")

# 設定圖片標題
plt.title("2021 Taipei Max/Min Temperature")

# 標示x軸
plt.xlabel("Month")
plt.xticks(month)
# 標示y軸
plt.ylabel("Temperature")
plt.yticks([y for y in range(0,40,3)])
# 顯示出線條標記位置
plt.legend(loc = "best")

# show
plt.grid(linestyle=':')
plt.show()
```
